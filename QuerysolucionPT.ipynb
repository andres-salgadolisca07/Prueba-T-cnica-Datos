```python
import pandas as pd
from pathlib import Path

parquet_path = Path("yellow_tripdata_2021-08.parquet")
```

```python
# Leer parquet completo (estable)
df = pd.read_parquet(parquet_path)

# Leer parquet
df = pd.read_parquet(parquet_path)
```

```python
df = df[[
    "tpep_pickup_datetime",
    "PULocationID"
]]

# Renonmbramos
df = df.rename(columns={"PULocationID": "vehicle_id"})

# No obtenidas de la data
df["ultima_latitud"] = pd.NA
df["ultima_longitud"] = pd.NA

# Asegurar datetime
df["tpep_pickup_datetime"] = pd.to_datetime(df["tpep_pickup_datetime"])

```

```python

# Salida 1: Ultima ubicaci√≥n

df_sorted = df.sort_values("tpep_pickup_datetime")

ultima_ubicacion = (
    df_sorted.groupby("vehicle_id", as_index=False)
    .last()
    .rename(columns={"tpep_pickup_datetime": "ultimo_timestamp"})
    [["vehicle_id", "ultimo_timestamp", "ultima_latitud", "ultima_longitud"]]
)

ultima_ubicacion.to_csv("ultima_ubicacion.csv", index=False)



print(" Archivo csv correctamente")
```

```python
# Salida 2 :Viajes por horas

df["hora_del_dia"] = df["tpep_pickup_datetime"].dt.hour

viajes_por_hora = (
    df.groupby("hora_del_dia")
    .size()
    .reset_index(name="total_viajes")
    .sort_values("hora_del_dia")
)

viajes_por_hora.to_csv("viajes_por_hora.csv", index=False)

print(" Archivo csv correctamente")
```
